```python
import psycopg2 
from psycopg2 import Error

try:
    # Connect to an existing database
    connection = psycopg2.connect(user="postgres",
                                  password="python",
                                  host="127.0.0.1",
                                  port="5432",
                                  database="test")

    # Create a cursor to perform database operations
    cursor = connection.cursor()
    sql = '''select id , emp_name,salary 
             from pytest.emps
             order by salary desc 
             limit 1 ''' 
    cursor.execute(sql)
    # Fetch result
    record = cursor.fetchone()
    
    for rec in record:
        print(rec)
    
except (Exception, Error) as error:
    print("Error while connecting to PostgreSQL", error)
finally:
    if (connection):
        cursor.close()
        connection.close()
        print("PostgreSQL connection is closed")
```

```python
import pandas as pd 

df = pd.DataFrame(record)
df
```

```python
import psycopg2 
from psycopg2 import Error

try:
    # Connect to an existing database
    connection = psycopg2.connect(user="postgres",
                                  password="python",
                                  host="127.0.0.1",
                                  port="5432",
                                  database="test")

    # Create a cursor to perform database operations
    cursor = connection.cursor()
  
    my_sql = ''' select e.emp_name,
                 tc.technology,e.salary,tc.experiance 
                 from pytest.emps e
                 inner join  pytest.my_tech tc
                 on tc.tech_id = e.tech
                 order by e.salary desc;
             '''
    cursor.execute(my_sql)
    # Fetch result
    record = cursor.fetchall()
    
    #for rec in record:
     #   print(rec)
    
except (Exception, Error) as error:
    print("Error while connecting to PostgreSQL", error)
finally:
    if (connection):
        cursor.close()
        connection.close()
        print("PostgreSQL connection is closed")
```
